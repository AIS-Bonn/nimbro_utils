## ColorPalette

```python
from nimbro_utils.lazy import ColorPalette
```

```python
my_palette = ColorPalette(
    colors={
        'black': "#000000",
        'white': "#FFFFFF",
        'red': "#FF0000",
        'green': "#00FF00",
        'blue': "#0000FF"
    },
    name="basic",
    groups={
        'color': ['red', 'green', 'blue'],
        'contrast': ['black', 'white']
    }
)
```

```python
print(my_palette)
```

```python
print("Name  :", my_palette.name)
print("Names :", my_palette.names)
print("Groups:", my_palette.groups)
print("HEX:          ", my_palette.hex)
print("HEX (shuffle):", my_palette.hex_shuffle)
print("RGB:          ", my_palette.rgb)
print("RGB (shuffle):", my_palette.rgb_shuffle)
print("BGR:          ", my_palette.bgr)
print("BGR (shuffle):", my_palette.bgr_shuffle)
```

```python
print(type(my_palette.black))
print(my_palette.black)
print(my_palette['black'])
```

```python
print(type(my_palette.contrast))
print(my_palette.color)
print(my_palette['color'])
```

```python
print("Num colors (total):   ", len(my_palette))
print("Num colors (color):   ", len(my_palette.color))
print("Num colors (contrast):", len(my_palette.contrast))
```

## Preset color palettes

```python
from nimbro_utils.lazy import nimbro, kelly, monokai, solarized, bonn, night, wave, scream, tangerine, globe, x11 # all of type ColorPalette
```

```python
for palette in [nimbro, kelly, monokai, solarized, bonn, night, wave, scream, tangerine, globe, x11]:
    print(f"Name: '{palette.name}' ({len(palette)})")
    if len(palette.groups.keys()):
        print(f"Groups: {list(palette.groups.keys())}")
    show_colors(palette)
```
